# Recommender System For a Medical Supplies Company

## Introduction

In the competitive market of medical supplies, providing personalized recommendations can significantly enhance customer satisfaction and drive incremental sales. The goal of this project is to develop an advanced recommender system for a medical supplies company to effectively promote products and increase sales by offering tailored suggestions based on customer preferences and purchasing behavior. Leveraging data analytics and machine learning techniques, this system will deliver relevant product recommendations, optimizing the customer experience and supporting strategic sales initiatives.

### Project Goals

The primary objective is to create a recommender system that boosts sales and improves product visibility by offering personalized recommendations to customers. The system will be built through a detailed process of data preparation and modeling, incorporating both simple and advanced recommendation strategies.

### Data Preparation and Modeling

1. **Handling Missing Data**: 
   - Address any gaps or inconsistencies in the dataset to ensure the reliability and accuracy of the analysis and the recommendations.

2. **Identifying Popular Products**:
   - Calculate and identify the most popular products based on maximum sales volume and total dollar amount. This will help to understand which products are currently in high demand and can be used as a baseline for recommendations.

3. **Top Customer Analysis**:
   - Determine the company or customer with the maximum purchases to gain insights into key buyer behavior and tailor recommendations to high-value segments.

4. **Popularity-Based Recommender System**:
   - Develop a recommendation system that suggests products based on overall popularity. This approach uses the sales volume and revenue data to recommend widely preferred products to both new and existing customers.

5. **Matrix Factorization for Recommendations**:
   - Implement a recommender system using matrix factorization techniques, such as Singular Value Decomposition (SVD) or Alternating Least Squares (ALS), to capture latent factors in customer-product interactions. This method allows for more nuanced and personalized recommendations based on underlying patterns in purchasing behavior.

6. **Cosine Similarity for Product Descriptions**:
   - Apply cosine similarity to analyze product descriptions, enabling the system to recommend similar products based on textual content. This enhances the relevance of recommendations, particularly for customers searching for specific product attributes or alternatives.

### Expected Outcomes

The development and implementation of this recommender system will provide several benefits:
- **Enhanced Customer Experience**: By delivering personalized product suggestions, customers will receive more relevant and useful recommendations, improving their overall shopping experience.
- **Increased Sales**: Targeted recommendations based on customer preferences and purchasing patterns will drive higher sales and promote more effective product upselling and cross-selling.
- **Actionable Insights**: The system will generate valuable insights into customer behavior and product performance, supporting more informed decision-making and strategic planning.

By integrating these advanced recommendation techniques, the project aims to create a robust and effective system that enhances both customer satisfaction and sales performance for the medical supplies company.

## Popularity-based recommender

```python
df["Products.cost"].fillna(df["Products.cost"].mean(),inplace=True)
```

```python
df["Products.cost"].isnull().sum()
```

```python
products_grouped=df.groupby("Products.name").agg({"Products.cost":[np.size,np.sum,np.mean]})
```

```python
products_grouped.head()
```

```python
popular_products=products_grouped.sort_values(("Products.cost","mean"),ascending=False)
```

```python
popular_products.head()
```

```python
grouped_sum=products_grouped["Products.cost"]["sum"].sum()
```

```python
popular_products["percentage"]=popular_products["Products.cost"]["sum"].div(grouped_sum)*100
```

```python
popular_products.head()
```

```python
popular_products=popular_products.sort_values("percentage",ascending=False)
```

```python
popular_products.head()
```

```python
popular_products["Rank"]=popular_products["percentage"].rank(ascending=False)
popular_products.head()
```

##  Cosine similarity for product description

```python
from textblob import TextBlob
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
pr=PorterStemmer()
def lemmafn(text):
    words=TextBlob(text).words
    return [pr.stem(word) for word in words]
df["Products.long_description"]=df["Products.long_description"].str.replace(r'<[^>]*>','',regex=True)
df["Products.long_description"]=df["Products.long_description"].str.lower()
df["Products.long_description"]=df["Products.long_description"].str.replace('[^\w\s]','',regex=True)
df["Products.long_description"]=df["Products.long_description"].str.replace('\d+','',regex=True)
df["Products.long_description"]=df["Products.long_description"].str.replace('\n','',regex=True)
vect=CountVectorizer(stop_words="english",ngram_range=(1,3),max_features=1000,analyzer=lemmafn)
```

```python
df["Products.long_description"].isnull().sum()
```

```python
df["Products.long_description"].fillna("No description available", inplace=True)
```

```python
matrix=vect.fit_transform(df["Products.long_description"])
```

```python
df=pd.concat([df,pd.DataFrame(matrix.toarray())])
```

```python
df.head()
```

```python
from scipy import spatial
def compute_distance(a,b):
    genre_distance=spatial.distance.cosine(a[1],b[1])
    popularity_distance=abs(a[2]-b[2])
    return genre_distance+popularity_distance
```

```python
product_dict={}
with open("data.csv",encoding="latin-1") as f:
    header = f.readline().strip().split(',')
    for line in f:
        fields=line.rstrip('\n').split(',')
        product_id=int(fields[83])
        name=fields[89]
        descriptions=fields[181:1180]
        descriptions=list(map(int,descriptions))
        product_dict[product_id]=[name,descriptions]
```

```python
distances=[]
for product1 in product_dict:
    for product2 in product_dict:
            distances.append(compute_distance(product_dict[product1],product_dict[product2]))
```
